```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)
```

```python
f, axarr = plt.subplots(1,2)

# column 1
axarr[0].plot(x,y)
axarr[0].set_xlabel("x")
axarr[0].set_ylabel("sin(x)")
axarr[0].set_title(r"$\sin(x)$")

# column 2
axarr[1].plot(x,z, color="red")
axarr[1].set_xlabel("x")
axarr[1].set_ylabel("cos(x)")
axarr[1].set_title(r"$\cos(x)$")
axarr[1].set_xlim([0, 2*np.pi])
axarr[1].set_ylim([-1,1])

f.subplots_adjust(wspace=0.5)
f.set_dpi(2000)
```

```python
f, axarr = plt.subplots(1, 2)

# column 1
axarr[0].plot(x, y)
axarr[0].set_xlabel("x")
axarr[0].set_ylabel("sin(x)")
axarr[0].set_title(r"$\sin(x)$")

# column 2
axarr[1].plot(x, z, color="red")
axarr[1].set_xlabel("x")
axarr[1].set_ylabel("cos(x)")
axarr[1].set_title(r"$\cos(x)$")

f.subplots_adjust(wspace=0.5)
f.set_dpi(2000)

axarr[0].set_aspect("equal")
axarr[1].set_aspect(np.pi)
```

```python
fig = plt.figure(figsize=(6,6))

plt.plot(x,y, label=r"$y = \sin(x)$")
plt.plot(x,z, label=r"$y = \cos(x)$")
plt.plot(x,w, label=r"$y = \sin(4x)$")
plt.plot(x,v, label=r"$y = \cos(4x)$")

plt.xlabel(r"$x$")
plt.ylabel(r"$y(x)$")
plt.xlim(0, 2*np.pi)
plt.ylim(-1.2, 1.2)
plt.gca().set_aspect(np.pi/1.2)
plt.legend(loc=1, frameon=True)
```
