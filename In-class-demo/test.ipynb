### Making multipanel plots with matplotlib

###### import numpy and matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

###### Define array of angles and their sins can cosines using numpy.

```python
# Create an array from 0 to 2pi, with 100 steps in between
x = np.linspace(0, 2 * np.pi, 100)
print(x[-1], 2 * np.pi)

y = np.sin(x)
z = np.cos(x)
w = np.sin(4 * x)
v = np.cos(4 * x)
```

##### Plot a sine curve

```python
# make a simple plot
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.xlim([0, 2 * np.pi])
plt.ylim([-1, 1])
```

```python
# make a simple plot
plt.plot(x, y, label = "sin(x)")
plt.plot(x, z, label = "cos(x)")
plt.plot(x, w, label = "sin(4x")
plt.plot(x, v, label = "cos(4x)")

plt.xlabel("x")
plt.ylabel("y, z, w, v")
plt.xlim([0, 2 * np.pi])
plt.ylim([-1.1, 1.1])
plt.legend()
```
